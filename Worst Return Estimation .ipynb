```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pickle
```

```python
with open('./logs/RtnEst_OT_radi_0.05/NaiveRtn.pickle', 'rb') as fp:
    OT_Naive = pickle.load(fp)
    
with open( './logs/RtnEst_OT_radi_0.05/sam.pickle', 'rb') as fp:
    OT_sam = pickle.load(fp)
    
OT_Est = np.mean(OT_sam[:, -10:], axis=1)

with open('./logs/RtnEst_SCOT_radi_0.05/NaiveRtn.pickle', 'rb') as fp:
    SCOT_Naive = pickle.load(fp)
    
with open( './logs/RtnEst_SCOT_radi_0.05/sam.pickle', 'rb') as fp:
    SCOT_sam = pickle.load(fp)
    
SCOT_Est = np.mean(SCOT_sam[:, -10:], axis=1)

plt.figure(figsize=(10, 6))
plt.plot(SCOT_Naive, '--o', markersize=4, label='Real Return')
plt.plot(SCOT_Est, '--o', markersize=4, label='Worst Return (SCOT)')
plt.plot(OT_Est, '--o', markersize=4, label='Worst Return (OT)')

# plt.xticks(np.linspace(0.955, 1.0, num=10), fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Time', fontsize = 12)
plt.ylabel('Return', fontsize = 12)
plt.legend(loc='best', fontsize=12)
plt.savefig('RtnEst_005.pdf', format='pdf', dpi=1000, bbox_inches='tight', pad_inches=0.1)
```
